This example checks for rows that are in New York OR California AND contain EITHER one of the items from skill_list OR one of the items in accessibility

```python
accessibility = ['collaborative', 'github', 'remote work','entry level', 'junior engineer', 'veteran']



location_condition = (df['Location'].str.contains(r'\bNew York\b|\bNY\b', case=False, na=False)) | (df['Location'].str.contains(r'\bCalifornia\b|\bCA\b', case=False, na=False))

skill_condition = df['Skills'].apply(lambda skills_in_row: any(skill in skills_in_row for skill in skill_list))

accessibility_condition = df['Accessibility'].apply(lambda acs_in_row: any(acs in acs_in_row for acs in accessibility))

multi_conditional_check = df.loc[(location_condition & skill_condition) | accessibility_condition]
multi_conditional_check.head()
```

And finally a function to input your own conditional arguments

```python
def filter_dataframe(df: pd.DataFrame, locs: list = None, skills: list = None, acs: list = None, index=df.index) -> pd.DataFrame:

    locs_condition = pd.Series(True, index=df.index)  # Initialize with all True
    list_skill_condition = pd.Series(True, index=df.index)
    acs_condition = pd.Series(True, index=df.index)

    if locs:
        location_patterns = []
        for loc in locs:
            loc = loc.strip()  # Remove leading/trailing whitespace
            if len(loc) == 2:  # Assume 2-letter entries are state abbreviations
                location_patterns.append(r'(?:^|, |\s)' + re.escape(loc) + r'(?:$|, |\s)')
            else:
                location_patterns.append(r'\b' + re.escape(loc) + r'\b')

        if location_patterns:
            combined_location_pattern = '|'.join(location_patterns)
            locs_condition = df['Location'].str.contains(combined_location_pattern, case=False, na=False)
        else:
            locs_condition = pd.Series(True, index=df.index)

    if skills:
        list_skill_condition = df['Skills'].apply(lambda skills_in_row: any(x in skills_in_row for x in skills))

    if acs:
        acs_condition = df['Accessibility'].apply(lambda acs_in_row: any(x in acs_in_row for x in acs))

    if locs and skills:
        combined_condition = (locs_condition & list_skill_condition) | acs_condition
    elif locs:
        combined_condition = locs_condition | acs_condition
    elif skills:
        combined_condition = list_skill_condition | acs_condition
    elif acs:
        combined_condition = acs_condition
    else:
        return df  # Return original DataFrame if no criteria are provided

    return df.loc[combined_condition]
```
